```python
demy=list(donnees['ACADEMIE'])
demy 
```

```python
liste=[]
def indique_total_devis_par_groupe(donnees,col,liste):
    total_devis_academy=[]

    for i in liste:
    
        aca=donnees[donnees[col]==i]
        devis=len(aca)
    
    
        total_devis_academy.append(devis)
    
    
    
    return total_devis_academy
```

```python
liste=[]
def indique_devis_accepte_par_groupe(donnees,col,liste):
    devis_accepte_academy=[]

    for i in liste:

        aca=donnees[donnees[col]==i]
        accepte=aca[aca['devis']=='Devis Accepte']
        nombre=len(accepte)
        devis_accepte_academy.append(nombre)


    return devis_accepte_academy 
```

```python
donnees['total devis par academy']=indique_total_devis_par_groupe(donnees,'ACADEMIE',demy)
donnees[' devis accepte par academy']=indique_devis_accepte_par_groupe(donnees,'ACADEMIE',demy)
donnees
```

```python
donnees[' academy accepte']=donnees[' devis accepte par academy']/ donnees['total devis par academy']

donnees
```

```python
destination=list(donnees['DESTINATION'])
 
donnees['total devis par destination']=indique_total_devis_par_groupe(donnees,'DESTINATION',destination)
donnees[' devis accepte par destination']=indique_devis_accepte_par_groupe(donnees,'DESTINATION',destination)
donnees
```

```python
donnees[' destination accepte']=donnees[' devis accepte par destination']/ donnees['total devis par destination']

donnees
```

```python
etablissement=list(donnees['ETABLISSEMENT'])
 
donnees['total devis par ETABLISSEMENT']=indique_total_devis_par_groupe(donnees,'ETABLISSEMENT',etablissement)
donnees[' devis accepte par ETABLISSEMENT']=indique_devis_accepte_par_groupe(donnees,'ETABLISSEMENT',etablissement)
donnees
```

```python
donnees[' ETABLISSEMENT accepte']=donnees[' devis accepte par ETABLISSEMENT']/ donnees['total devis par ETABLISSEMENT']

donnees
```
